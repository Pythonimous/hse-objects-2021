```python
import pandas as pd
import numpy as np
```

```python
import os
from tqdm.notebook import tqdm
```

```python
os.chdir('gdrive/MyDrive/yelp_task/yelp_data')
```

```python
br_train = pd.read_csv('business_restaurant/train.csv')
br_dev = pd.read_csv('business_restaurant/dev.csv')
br_test = pd.read_csv('business_restaurant/test.csv')
```

```python
X_train_files, y_train = list(br_train.photo_id) + list(br_dev.photo_id), list(br_train.restaurant) + list(br_dev.restaurant)
X_test_files, y_test = list(br_test.photo_id), list(br_test.restaurant)
```

```python
X_train, X_test = [], []
```

```python
features = np.load('transfer_features/alexnet_features.npz', allow_pickle = True)['arr_0'][()]
```

```python
for filename in tqdm(X_train_files):
    X_train.append(features[filename])
```

```python
for filename in tqdm(X_test_files):
    X_test.append(features[filename])
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
X_test.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix, roc_auc_score
```

```python
clf = make_pipeline(StandardScaler(), BernoulliNB())
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
confusion_matrix(y_test, clf.predict(X_test))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, clf.predict(X_test)).ravel()
print(f"tn: {tn}, fp: {fp}, fn: {fn}, tp: {tp}")
```

```python
roc_auc_score(y_test, clf.predict(X_test))
```
